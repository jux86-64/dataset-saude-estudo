```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Tema 1. Evolução dos Casos de Dengue nos Estados do Sudeste (2015–2024)


#### Objetivo: Analisar a evolução dos casos de dengue, identificando sazonalidade, picos epidêmicos e diferenças entre estados.

```python
url = '/content/drive/MyDrive/Colab Notebooks/datasets/Simulacao_Dengue_Sudeste_2015_2024.xlsx'
df = pd.read_excel(url)
```

```python
print(df)
```

```python
agrupar = df.groupby(['Ano', 'Estado']).sum()
print(agrupar)
```

```python
# Grafico do Agrupamento

plt.figure(figsize=(12, 6))

# data=df -> usa o dataframe
# x='Ano' -> Coloca a coluna 'Anos' no eixo X
# y='Casos Confirmados' -> Coloca a coluna 'Casos Confirmaods' no eixo y
# hue='Estado -> Cria barras sepadaradas para a coluna 'Estado'
sns.barplot(data=df, x='Ano', y='Casos Confirmados', hue='Estado')

plt.title('Agrupamento de Casos por Ano e Estado')
plt.xlabel('Ano')
plt.ylabel('Número de Casos')

# bbox_to_anchor=(1.05, 1) -> Posiciona a legenda FORA do gráfico à direita
# loc='upper left' -> Alinha a legenda no canto superior esquerdo do anchor
plt.legend(title='Estado', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
ano_maior_casos = df.groupby('Ano')['Casos Confirmados'].sum()

print(ano_maior_casos)
```

```python
# Gráfico do ano com maior número de casos

plt.bar(df['Ano'], df['Casos Confirmados'])
plt.title('Ano com maior número de casos confirmados')
plt.xlabel('Ano')
plt.ylabel('Casos Confirmados')
plt.ylim(50000,100000)
plt.xticks(df['Ano'])
plt.show()
```

```python
estado_mais_afetado = df.groupby('Estado')['Casos Confirmados'].sum()

print(estado_mais_afetado)
```

```python
# Gráfico de estado mais afetado

plt.bar(estado_mais_afetado.index, estado_mais_afetado.values)
plt.title('Estado mais afetado pela dengue')
plt.xlabel('Estado')
plt.ylabel('Número de Casos')
plt.ylim(450000,650000)
plt.tight_layout()
plt.show()
```
